### Part 1: Cross validation without parameter tuning

```python
## Setting the 10 fold stratified cross-validation
cvKFold=StratifiedKFold(n_splits=10, shuffle=True, random_state=0)

# The stratified folds from cvKFold should be provided to the classifiers
```

```python
from sklearn.metrics import accuracy_score
# K-Nearest Neighbour
from sklearn.neighbors import KNeighborsClassifier

def kNNClassifier(X, y, k):
  scores=[]
  knn = KNeighborsClassifier(n_neighbors=k)
  for a,b in cvKFold.split(X, y):
    x_train, x_test = X.iloc[a], X.iloc[b]
    y_train, y_test = y.iloc[a], y.iloc[b] 
    knn.fit(x_train, y_train)
    y_pred = knn.predict(x_test) 
    scores.append(accuracy_score(y_pred, y_test))
  scores = np.array(scores)
  return scores.mean()
```

```python
# Logistic Regression
from sklearn.linear_model import LogisticRegression
def logregClassifier(X, y):
  scores=[]
  log = LogisticRegression(random_state=0)
  for a,b in cvKFold.split(X, y):
    x_train, x_test = X.iloc[a], X.iloc[b]
    y_train, y_test = y.iloc[a], y.iloc[b] 
    log.fit(x_train, y_train)
    y_pred = log.predict(x_test) 
    scores.append(accuracy_score(y_pred, y_test))
  scores = np.array(scores)
  return scores.mean()
```

```python
#Naïve Bayes
from sklearn.naive_bayes import GaussianNB
def nbClassifier(X, y):
  scores=[]
  nb = GaussianNB()
  for a,b in cvKFold.split(X, y):
    x_train, x_test = X.iloc[a], X.iloc[b]
    y_train, y_test = y.iloc[a], y.iloc[b] 
    nb.fit(x_train, y_train)
    y_pred = nb.predict(x_test) 
    scores.append(accuracy_score(y_pred, y_test))
  scores = np.array(scores)
  return scores.mean()
```

```python
# Decision Tree
from sklearn.tree import DecisionTreeClassifier
def dtClassifier(X, y):
  scores=[]
  tree = DecisionTreeClassifier(criterion='entropy',random_state=0)
  for a,b in cvKFold.split(X, y):
    x_train, x_test = X.iloc[a], X.iloc[b]
    y_train, y_test = y.iloc[a], y.iloc[b]  
    tree.fit(x_train, y_train)
    y_pred = tree.predict(x_test) 
    scores.append(accuracy_score(y_pred, y_test))  
  scores = np.array(scores)
  return scores.mean()
```

```python
# Ensembles: Bagging, Ada Boost and Gradient Boosting
from sklearn.ensemble import BaggingClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import AdaBoostClassifier
from sklearn.ensemble import GradientBoostingClassifier

def bagDTClassifier(X, y, n_estimators, max_samples, max_depth):
  scores=[]
  bag = BaggingClassifier(DecisionTreeClassifier(criterion='entropy',  max_depth=max_depth ,random_state=0), n_estimators=n_estimators, max_samples=max_samples, bootstrap=True, random_state=0)
  for a,b in cvKFold.split(X, y):
      x_train, x_test = X.iloc[a], X.iloc[b]
      y_train, y_test = y.iloc[a], y.iloc[b]  
      bag.fit(x_train, y_train)
      y_pred = bag.predict(x_test)
      scores.append(accuracy_score(y_pred, y_test))
  scores = np.array(scores)
  return scores.mean()

def adaDTClassifier(X, y, n_estimators, learning_rate, max_depth):
    scores=[]
    ada = AdaBoostClassifier(DecisionTreeClassifier(criterion='entropy', max_depth=max_depth), n_estimators=n_estimators, learning_rate=learning_rate, random_state=0)
    for a,b in cvKFold.split(X, y):
      x_train, x_test = X.iloc[a], X.iloc[b]
      y_train, y_test = y.iloc[a], y.iloc[b]  
      ada.fit(x_train, y_train)
      y_pred = ada.predict(x_test)
      scores.append(accuracy_score(y_pred, y_test))
    scores = np.array(scores)
    return scores.mean()

def gbClassifier(X, y, n_estimators, learning_rate):
  scores=[]
  gb = GradientBoostingClassifier( n_estimators=n_estimators, learning_rate=learning_rate, random_state=0)
  for a,b in cvKFold.split(X, y):
      x_train, x_test = X.iloc[a], X.iloc[b]
      y_train, y_test = y.iloc[a], y.iloc[b]  
      gb.fit(x_train, y_train)
      y_pred = gb.predict(x_test)
      scores.append(accuracy_score(y_pred, y_test))
  scores = np.array(scores)
  return scores.mean()
```

```python
#Creating the training and class column.
class_col=new_data["class"]
new_data.drop("class", axis=1, inplace=True)
X= new_data
#KNN
k=3

#Bagging
bag_n_estimators = 50
bag_max_samples = 100
bag_max_depth = 5

#AdaBoost
ada_n_estimators = 50
ada_learning_rate = 0.5
ada_bag_max_depth = 5

#GB
gb_n_estimators = 50
gb_learning_rate = 0.5

# Print results for each classifier in part 1 to 4 decimal places here:
print("kNN average cross-validation accuracy: %.4f"%(kNNClassifier(X, class_col, k)))
print("LR average cross-validation accuracy: %.4f"%(logregClassifier(X,class_col)))
print("NB average cross-validation accuracy: %.4f"%(nbClassifier(X,class_col)))
print("DT average cross-validation accuracy: %.4f"%(dtClassifier(X, class_col)))
print("Bagging average cross-validation accuracy: %.4f"%(bagDTClassifier(X, class_col, bag_n_estimators, bag_max_samples, bag_max_depth)))
print("AdaBoost average cross-validation accuracy: %.4f"%(adaDTClassifier(X, class_col, ada_n_estimators, ada_learning_rate, ada_bag_max_depth)))
print("GB average cross-validation accuracy: %.4f"%(gbClassifier(X, class_col, gb_n_estimators, gb_learning_rate)))
```

### Part 2: Cross validation with parameter tuning

```python

Y = class_col
from sklearn.model_selection import train_test_split
X_train, X_test, Y_train, Y_test = train_test_split(X, Y,
                                                    stratify=Y, 
                                                     random_state=0)
```

```python
# Linear SVM
# You should use SVC from sklearn.svm
from sklearn.svm import SVC
from sklearn.model_selection import GridSearchCV


def bestLinClassifier(X,y):
  linear_svm=SVC(kernel="linear",random_state=0)
  grid_search = GridSearchCV(linear_svm,grid,cv=cvKFold)
  grid_search.fit(X,y)
  test_score=grid_search.score(X_test,Y_test)
  return  [grid_search.best_params_,grid_search.best_score_,test_score]

```

```python
# Random Forest
# You should use RandomForestClassifier from sklearn.ensemble with information gain and max_features set to ‘sqrt’.
from sklearn.ensemble import RandomForestClassifier

def bestRFClassifier(X,y):
  myclf = RandomForestClassifier(criterion='entropy',max_features = 'sqrt',random_state=0)
  grid_search = GridSearchCV(myclf,grid,cv=cvKFold)
  grid_search.fit(X,y)
  test_score=grid_search.score(X_test,Y_test)
  return  [grid_search.best_params_,grid_search.best_score_,test_score]
```
